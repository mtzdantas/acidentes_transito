```python
import pandas as pd
```

### Carregar e inspecionar o dataset

```python
acidentes = pd.read_csv('./data/accidents_2017_to_2023.csv')
```

```python
# Mostrando todas as colunas.
pd.set_option('display.max_columns', None)

# Visualizando tabela.
acidentes.head()
```

```python
# Conferindo tipos de dados das colunas.
acidentes.dtypes
```

### Eliminar colunas que não serão utilizadas

```python
# Excluindo colunas que não serão utilizadas.
acidentes = acidentes.drop(columns=['br', 'km', 'regional', 'delegacia'])
```

### Tratar valores ausentes (NaN)

```python
# Conferindo valores ausentes.
acidentes.isna().sum()
```

```python
# Substituindo valores ausentes da coluna tipo_acidente por desconhecido.
acidentes['tipo_acidente'] = acidentes['tipo_acidente'].fillna('Desconhecido')
```

```python
# Visualizando tabela.
acidentes[acidentes['tipo_acidente'] == 'Desconhecido'].head()
```

### Corrigir tipos de dados

```python
# Transformando a coluna de data para o tipo datetime.
acidentes['data_inversa'] = pd.to_datetime(acidentes['data_inversa'], format='%Y-%m-%d')

# Separando a data em colunas de dia, mês e ano.
acidentes['dia'] = acidentes['data_inversa'].dt.day
acidentes['mês'] = acidentes['data_inversa'].dt.month
acidentes['ano'] = acidentes['data_inversa'].dt.year
acidentes['nome_mês'] = acidentes['data_inversa'].dt.strftime('%B')

# Transformando nome_mês e dia_semana como tipo categórico para ordenação.
ordem_meses = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
ordem_semana = ['domingo', 'segunda-feira', 'terça-feira', 'quarta-feira', 'quinta-feira', 'sexta-feira', 'sábado'] 

acidentes['nome_mês'] = pd.Categorical(acidentes['nome_mês'], categories=ordem_meses, ordered=True)
acidentes['dia_semana'] = pd.Categorical(acidentes['dia_semana'], categories=ordem_semana, ordered=True)

# Excluindo coluna.
acidentes = acidentes.drop(columns=['data_inversa'])
```

```python
# Visualizando tabela.
acidentes.head()
```

### Remover valores duplicados

```python
# Conferindo se existe linhas duplicadas.
acidentes.duplicated().sum()
```

```python
# Removendo linhas duplicadas.
acidentes = acidentes.drop_duplicates()
```

### Análise Temporal


```python
# Agrupando quantidade de acidentes por mês.
acidentes_mes = acidentes.groupby('nome_mês', observed=False)['nome_mês'].count().reset_index(name='quantidade')
acidentes_mes
```

```python
# Agrupando quantidade de acidentes pelo dia da semana.
acidentes_semana = acidentes.groupby('dia_semana', observed=False)['dia_semana'].count().reset_index(name='quantidade')
acidentes_semana
```

### Análise de Causa de Acidente

```python
# Listando causas de acidentes que ultrapassem o 1% dos acidentes.
total_acidentes = len(acidentes)
principais_causas = acidentes['causa_acidente'].value_counts()
principais_causas = principais_causas[principais_causas > total_acidentes * 0.01].reset_index()
principais_causas
```

```python
# Filtrando causas de acidentes relacionados a Álcool, Sono e Animais na Pista que provavelmente ocorrem mais a noite.
causas = principais_causas[principais_causas['causa_acidente'].isin(['Ingestão de Álcool', 'Ingestão de álcool pelo condutor', 'Condutor Dormindo', 'Animais na Pista'])]

# Juntando linhas que representavam informações iguais com nomes diferentes.
causas.loc[causas['causa_acidente'] == 'Ingestão de álcool pelo condutor', 'causa_acidente'] = 'Ingestão de Álcool'
causas = causas.groupby('causa_acidente')['count'].sum().sort_values(ascending=False).reset_index()

causas
```

```python
# Comparando a diferença na quantidade de acidentes de cada causa conforme a fase do dia.
causas_fase_dia = pd.merge(causas[['causa_acidente']], acidentes[['causa_acidente', 'fase_dia']], on='causa_acidente')
causas_fase_dia = causas_fase_dia.groupby(['causa_acidente', 'fase_dia'])['fase_dia'].count().reset_index(name='quantidade')
causas_fase_dia
```

### Análise por Localização

```python
# Agrupando quantidade de acidentes por estado.
acidentes_estado = acidentes.groupby('uf', observed=False)['uf'].count().sort_values(ascending=False).reset_index(name='quantidade')
acidentes_estado
```
